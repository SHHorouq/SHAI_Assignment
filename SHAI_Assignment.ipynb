#Descriptive Statistics

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy

Salary = df['TotalPay']
print("First Way")
print(df['TotalPay'].describe())
print()

print("second way")
mean = np.mean(Salary)
median = np.median(Salary)
mode = scipy.stats.mode(Salary)
minimum = np.minimum(Salary[0],Salary[148653])
maximum = np.maximum(Salary[0],Salary[148653])
variance = np.var(Salary)
std_dev = np.std(Salary)
rang = maximum - minimum

#Calculate basic statistics mean, median, mode, minimum, and maximum salary, determine the range of salaries, and find the standard deviation
print("Mean= ", mean)
print("Median= ", median)
print("Mode= ", mode)
print("minimum= ", minimum)
print("maximum= ", maximum)
print("variance= ", variance)
print("range= ", rang)
print("standard deviation= ", std_dev)
```

#Data Cleaning: Handle missing data by suitable method with explain why you use it.

```python
#First way: Dropping irrelevant columns that logically don’t have any values:
df.drop(['Notes', 'Status'], axis=1, inplace=True)
print("the number of columns after dropping: ", df.shape[1])

#Second way: Handling missing data by imputing missing values with the median
df['OvertimePay'].fillna(df['OvertimePay'].median(), inplace=True)
df['OtherPay'].fillna(df['OtherPay'].mean(), inplace=True)

#third Way: Handling missing data by removing rows with null
df.dropna(subset=['BasePay'], inplace=True)
df.dropna(subset=['Benefits'], inplace=True)
print(df.shape)
print()
counts = df.isna().sum()
print(counts)
print()
```

#Basic Data Visualization

```python
import matplotlib.pyplot as plt

#Create histograms or bar charts to visualize the distribution of salaries.
Bens = [1000,100000,200000,300000]
plt.figure(figsize=(10,5))
plt.hist(df.TotalPay, bins=Bens)
plt.xticks(Bens)
plt.xlabel('TotalPay')
plt.ylabel('Number of Employees')
plt.title('distribution of salaries based on Number of Employees')
plt.show()
```
